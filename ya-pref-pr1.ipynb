```python
for i in honest_score['platform'].unique():
    plat_cor = honest_score.loc[:, 'general_sales' : 'user_score'].corr()
    print('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è {plat}:\n {mat_cor}\n'.format(plat=i, mat_cor=plat_cor))
    count = 0
    for j in plat_cor['general_sales'].unique():
        if j != 1:          
            if j >= 0.5:
                if count == 0:
                    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ - –∑–∞–º–µ—Ç–Ω–∞—è')
                elif count == 1:
                    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∑—Ä–∏—Ç–µ–ª–µ–π - –∑–∞–º–µ—Ç–Ω–∞—è')
            elif j >= 0.7:
                if count == 0:
                    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ - –≤—ã—Å–æ–∫–∞—è')
                elif count == 1:
                    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∑—Ä–∏—Ç–µ–ª–µ–π - –≤—ã—Å–æ–∫–∞—è')
            else:
                if count == 0:
                    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ - –Ω–∏–∑–∫–∞—è')
                elif count == 1:
                    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∑—Ä–∏—Ç–µ–ª–µ–π - –Ω–∏–∑–∫–∞—è')
            count += 1 
      
    plat_stat = honest_score[honest_score['platform'] == i]
    
    plat_stat.plot(x='critic_score', y='general_sales', kind='scatter')
    plt.title('–í–ª–∏—è–Ω–∏–µ –æ—Ç–∑—ã–≤–æ–≤ –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ {plat}'.format(plat=i))
    plt.xlabel('–û—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤')
    plt.ylabel('–û–±—â–∏–µ –ø—Ä–æ–¥–∞–∂–∏')
    plt.show()
    
    plat_stat.plot(x='user_score', y='general_sales', kind='scatter')
    plt.title('–í–ª–∏—è–Ω–∏–µ –æ—Ç–∑—ã–≤–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ {plat}'.format(plat=i))
    plt.xlabel('–û—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
    plt.ylabel('–û–±—â–∏–µ –ø—Ä–æ–¥–∞–∂–∏')
    plt.show()
```

**–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0)**: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Å–æ–≤–ø–∞–¥–∞—é—Ç.

**–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1)**: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç.

```python
alpha = .05

h0 = current_period[current_period['genre'] == 'Action']['user_score'].dropna()
h1 = current_period[current_period['genre'] == 'Sports']['user_score'].dropna()

results = st.ttest_ind(
    h0,
    h1,
    equal_var=True
)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

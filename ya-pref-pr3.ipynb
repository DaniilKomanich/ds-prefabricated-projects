```python
def kdeplot(df, figsize, x, y, suptitle, hue = None):
    fig, ax = plt.subplots(figsize = figsize)
    fig.suptitle(suptitle, fontsize = 14)
    kdeplt = sns.kdeplot(df, hue = hue) 
    kdeplt.grid(color = 'lightblue', linewidth = 1, linestyle=':')
    kdeplt.set_xlabel(x, fontsize = 12)
    kdeplt.set_ylabel(y, fontsize = 12)
    return kdeplt
```

```python
kdeplot(df['ServiceDuration'], 
        figsize = (15, 5), 
        x = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è: 1 = —É—à–µ–ª, 0 = –ø–æ–ª—å–∑—É–µ—Ç—Å—è',
        y = '–ü–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è',
        suptitle = '–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–Ω–µ–π –ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —É—Å–ª—É–≥–∞–º–∏',
        hue = df['CurrentUser'])
plt.show()
```

#### `LogisticRegression`

```python
start = time.time()

param_grid = {
    'C': range(1, 5, 1),
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'max_iter': range(150, 251, 50),
    'tol': [0.001, 0.01, 0.1]
}

model = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced', n_jobs=-1)
model_lr = GridSearchCV(model, param_grid, cv=5, scoring='roc_auc')
model_lr.fit(features_train_lr, target_train)

best_params_lr = model_lr.best_params_
print('\n–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è LogisticRegression', best_params_lr)
auc_train_lr = model_lr.best_score_
print('ROC-AUC –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', auc_train_lr)

lr_time_train = time.time() - start 
print('\n–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —è—á–µ–π–∫–∏:', round(lr_time_train/60), '–º–∏–Ω')
```

#### `RandomForrestClassifier`

```python
start = time.time()

param_grid = {
    'max_depth': range(5, 16, 5),
    'min_samples_split': range(1, 5, 1),
    'min_samples_leaf': range(5, 21, 2),
    'n_estimators': [50, 100, 150], 
}

model = RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced')
model_rf = GridSearchCV(model, param_grid, cv=5, scoring='roc_auc')
model_rf.fit(features_train, target_train)

best_params_rf = model_rf.best_params_
print('\n–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è RandomForestClassifier', best_params_rf)
auc_train_rf = model_rf.best_score_
print('ROC-AUC –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', auc_train_rf)

rf_time_train = time.time() - start 
print('\n–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —è—á–µ–π–∫–∏:', round((rf_time_train/60), 2), '–º–∏–Ω')
```

#### `CatBoostClassifier`

```python
start = time.time()

param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'iterations': range(800, 1001, 100),
    'depth': range(1, 12, 5),
    'l2_leaf_reg': [1, 3, 5]
}

model = CatBoostClassifier(random_state=RANDOM_STATE, verbose=False)
model_cb = GridSearchCV(model, param_grid, cv=5, scoring='roc_auc')
model_cb.fit(features_train, target_train)

best_params_cb = model_cb.best_params_
print('\n–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è CatBoostClassifier', best_params_cb)
auc_train_cb = model_cb.best_score_
print('ROC-AUC –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', auc_train_cb)

cb_time_train = time.time() - start 
print('\n–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —è—á–µ–π–∫–∏:', round((cb_time_train/60), 2), '–º–∏–Ω')
```
